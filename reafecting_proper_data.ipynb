```python
# function to count how many pictures in a folder
def count_pictures_in_folder(folder_path: str) -> int:
    count = 0
    for file in os.listdir(folder_path):
        if file.endswith(".webp"):
            count += 1
    return count
    # function to count how many folders in a directory
def count_folders_in_directory(directory_path: str) -> int:
    count = 0
    for file in os.listdir(directory_path):
        if os.path.isdir(os.path.join(directory_path, file)):
            count += 1
    return count

count_pictures_in_folder("/workspaces/AICoinXpert/algo/webscraping/data/selected_coins_above20")
```

```python
import os

def get_folder_names(path):
    """
    Get all the names of the folders and files in a directory.

    Args:
        path (str): The path of the directory.

    Returns:
        list: A list containing the names of the folders and files.
    """
    return [name for name in os.listdir(path)]

folder_name = get_folder_names("/workspaces/AICoinXpert/algo/webscraping/data/organized_images_20_above_filtered/train")
len(folder_name)
```

```python
import os
import pandas as pd

def compare_and_update_dataframe(df, directory_path, min_matching_blocks_range=(4, 5)):
    for index, row in df.iterrows():
        folder_path_parts = row['folder_path'].split('-')
        matching_file_path = None
        
        common_parts = set()  # Initialize the common parts to an empty set
        
        for file_name in os.listdir(directory_path):
            if file_name.endswith('.webp'):
                file_name_parts = file_name.split('-')
                common_parts = set(folder_path_parts) & set(file_name_parts)
            
            if len(common_parts) in range(min_matching_blocks_range[0], min_matching_blocks_range[1]+1):
                matching_file_path = os.path.join(directory_path, file_name)
                break
        
        if matching_file_path:
            df.at[index, 'matching_file_path'] = matching_file_path
        else:
            df.at[index, 'matching_file_path'] = None
    
    return df

directory_path = '/workspaces/AICoinXpert/algo/webscraping/data/selected_coins_above20'

df = compare_and_update_dataframe(df, directory_path)
df
```

```python
# function to get the average tirage of coins values per country
def fill_tirage_with_average(df):
    df['tirage'] = pd.to_numeric(df['tirage'], errors='coerce').astype(float)  # Convert 'tirage' column to float and handle NaN values
    
    for country in df['country'].unique():
        mean_tirage = df[df['country'] == country]['tirage'].mean()
        df.loc[(df['country'] == country) & (df['tirage'].isnull()), 'tirage'] = mean_tirage
        
    return df

df_before_post = fill_tirage_with_average(df_before_post)
```
